```python
import pandas as pd

# Example DataFrame creation (use your actual DataFrame)
# df = pd.DataFrame({ ... })

common_outlier_indices = None  # Initialize a variable to store common indices

for col in df.columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Get the indices of outliers
    outlier_indices = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
    
    print(f"Column: {col} - Number of Outliers: {len(outlier_indices)}")
    
    if common_outlier_indices is None:
        common_outlier_indices = set(outlier_indices)  # Initialize with the first column's outliers
    else:
        common_outlier_indices.intersection_update(outlier_indices)  # Keep only common indices

# Convert to a list if needed
common_outlier_indices = list(common_outlier_indices)

# Display the common outlier indices
print(f"Common Outlier Indices: {common_outlier_indices}")
```

```python
x = df_transformed.drop(columns=['grip_lost'])  # Features
y = df_transformed['grip_lost'] 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
len(xtrain),len(ytrain),len(xtest),len(ytest)
```

```python
def mymodel(model):
    # Fit the model
    model.fit(xtrain, ytrain)
    
    # Predictions
    ypred = model.predict(xtest)
    xpred = model.predict(xtrain)
    
    # Printing predictions
    print("Predictions on Test Set:", ypred)
    print("Predictions on Train Set:", xpred)
    
    # Accuracy scores
    test_ac = accuracy_score(ytest, ypred)
    print(f"Test Accuracy: {test_ac:.4f}")  # Format to 4 decimal places
    train_ac = accuracy_score(ytrain, xpred)
    print(f"Train Accuracy: {train_ac:.4f}")  # Format to 4 decimal places
    
    # Confusion matrix
    cm = confusion_matrix(ytest, ypred)
    print("Confusion Matrix:")
    print(cm)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', cbar=False, xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
    
    # Classification report
    report = classification_report(ytest, ypred)
    print("Classification Report:")
    print(report)
    
    return model

# Initialize the logistic regression model
logreg = LogisticRegression()

# Call the model function
res = mymodel(logreg)
```
